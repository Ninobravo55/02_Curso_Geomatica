## Determinar indices espectrales RGB

```r
cal_indices_RGB <- function(B, G, R, N, 
                            indices = c("NGRDI","ExG","ExR","ExB","IKAW","GRRI","ExGR","GBDI","MGRVI","RGBVI","VDVI",
                                        "VARI","RGRI","NGBDI","CIVE","indices total")){
    # (Han etal., 2019)
    if(indices=="NGRDI"){
        cal_indices <- (G - R) / (G + R)
    }
    # (Han etal., 2019)
    if(indices=="ExG"){
        cal_indices <- (2* G) - R - B
    }
    # (Bendig etal., 2015)
    if(indices=="ExR"){
        cal_indices <- (1.4*R) - G
    }
    # (Ballesteros etal., 2018; Guo etal., 2020)
    if(indices=="ExB"){
        cal_indices <- (1.4*B) - G
    }
    # (Du & Noguchi, 2017)
    if(indices=="IKAW"){
        cal_indices <- (R - B)/(R + B)
    }
    # (Beniaich etal., 2019)
    if(indices=="GRRI"){
        cal_indices <- G/R
    }
    # (Lussem etal., 2018)
    if(indices=="ExGR"){
        cal_indices <- ExG - ExR
    }
    # # (Bendig etal., 2015; Lussem etal., 2018)
    if(indices=="GBDI"){
        cal_indices <- G-B
    }
    # (Wan etal., 2018)
    if(indices=="MGRVI"){
        cal_indices <- (G * G - R * R)/(G * G * R * R)
    }
    # (Lussem etal., 2018)
    if(indices=="RGBVI"){
        cal_indices <- (G * G - B * R)/(G * G + B * R)
    }
    # (Wan etal., 2018)
    if(indices=="VDVI"){
        cal_indices <- (2 * G - R - B)/(2 * G + R +B)
    }
    # # ndice De Resistencia Atmosfricamente Visible (VARI) (Beniaich etal., 2019)
    if(indices=="VARI"){
        cal_indices <- (G - R)/(G + R - B)
    }
    # 
    if(indices=="RGRI"){
        cal_indices <- R/G
    }
    # 
    if(indices=="NGBDI"){
        cal_indices <- (G - B)/(G + B)
    }
    # 
    if(indices=="CIVE"){
        cal_indices <- (0.441 * R - 0.8818 * G + 0.385 * B + 18.787)
    }
    if(indices=="indices total"){
        # Determinar los indices
        NGRDI <- (G - R) / (G + R)
        ExG <- (2* G) - R - B
        ExR <- (1.4*R) - G
        ExB <- (1.4*B) - G
        IKAW <- (R - B)/(R + B)
        GRRI <- G/R
        ExGR <- ExG - ExR
        GBDI <- G-B
        MGRVI <- (G * G - R * R)/(G * G * R * R)
        RGBVI <- (G * G - B * R)/(G * G + B * R)
        VDVI <- (2 * G - R - B)/(2 * G + R +B)
        VARI <- (G - R)/(G + R - B)
        RGRI <- R/G
        NGBDI <- (G - B)/(G + B)
        CIVE <- (0.441 * R - 0.8818 * G + 0.385 * B + 18.787)
        cal_indices_RGB <- stack(NGRDI,ExG,ExR,ExB, IKAW,GRRI,ExGR,GBDI,MGRVI,RGBVI,
                                 VDVI,VARI,RGRI,NGBDI,CIVE)
    }
    return(cal_indices_RGB)
}
```

## Determinar indices espectrales NIR

```r
cal_indices_NIR <- function(BLUE, GRE, RED, NIR, 
                        indices = c("NDVI","EVI","SAVI","MSAVI","NDWI","GNDVI","OSAVI","ARVI","SIPI","CGI",
                                    "RVI","CVI","CCCI","RECI","SVI","SSI","SCI","SMI","indices total")){
    # ndice de Vegetacin de Diferencia Normalizada (NDVI)
    if(indices=="NDVI"){
        cal_indices <- (NIR - RED) / (NIR + RED)
    }
    # ndice de Vegetacin de Diferencia Normalizada Mejorada (EVI)
    if(indices=="EVI"){
        cal_indices <- 2.5*(NIR - RED)/(NIR + 6*RED + 7.5*BLUE + 1)
    }
    # ndice De Vegetacin Ajustado Al Suelo (SAVI)
    if(indices=="SAVI"){
        cal_indices <- ((NIR - RED)/(NIR + RED + 0.5))*(1+0.5)
    }
    # ndice De Vegetacin Ajustado Al Suelo Modificado (MSAVI)
    if(indices=="MSAVI"){
        cal_indices <- (2*NIR + 1 - sqrt((2 * NIR + 1)**2 - 8*(NIR - RED)))/2
    }
    # ndice De Agua De Diferencia Normalizada (NDWI)
    if(indices=="NDWI"){
        cal_indices <- (GRE - NIR)/(GRE + NIR)
    }
    # ndice De Vegetacin De Diferencia Normalizada Verde (GNDVI)
    if (indices=="GNDVI"){
        cal_indices <- (NIR - GRE)/(NIR + GRE)
    }
    # ndice De Vegetacin Ajustado Al Suelo Optimizado (OSAVI)
    if (indices=="OSAVI"){
        cal_indices <- (NIR - RED) / (NIR + RED + 0.16)
    }
    # ndice De Vegetacin Resistente A La Atmsfera (ARVI)
    if (indices=="ARVI"){
        cal_indices <- (NIR - (2 * RED) + BLUE) / (NIR + (2 * RED) + BLUE)
    }
    # ndice De Pigmentacin Insensible A La Estructura (SIPI)
    if (indices=="SIPI"){
        cal_indices <- (NIR - BLUE) / (NIR - RED)
    }
    # ndice De Clorofila Verde (CGI)
    if (indices=="CGI"){
        cal_indices <- (NIR/GRE) - 1
    }
    # Relacin ndice de vegetacin
    if (indices=="RVI"){
        cal_indices <- (NIR/RED)
    }
    # Indice de Vegetacin de Clorofila (CVI)
    if (indices=="CVI"){
        cal_indices <- (NIR*RED/(GRE*GRE))
    }
    # ndice de contenido de clorofila del dosel
    if (indices=="CCCI"){
        cal_indices <- ((NIR / RED) - 1) / ((NIR / RED) + 1 + BLUE)
    }
    # ndice De Clorofila De Borde Rojo (ReCl)
    if(indices=="RECI"){
        cal_indices <- (NIR / RED) - 1
    }
    # ndice de Vegetacin del Suelo
    if(indices=="SVI"){
        cal_indices <- NIR / (RED + GRE)
    }
    # ndice de Salinidad del Suelo (SSI)
    if(indices=="SSI"){
        cal_indices <- (NIR / RED) - 1
    }
    # ndice de Carbono del Suelo (SCI)
    if(indices=="SCI"){
        cal_indices <- (NIR + RED)/(NIR * RED)
    }
    # ndice de Humedad del Suelo (SMI):
    if(indices=="SMI"){
        cal_indices <- (NIR / RED) ** 0.5
    }
    if(indices=="indices total"){
        # Determinar los indices
        NDVI <- (NIR - RED) / (NIR + RED)
        EVI <- 2.5*(NIR - RED)/(NIR + 6*RED + 7.5*BLUE + 1)
        SAVI <- ((NIR - RED)/(NIR + RED + 0.5))*(1+0.5)
        MSAVI <-(2*NIR + 1 - sqrt((2 * NIR + 1)**2 - 8 * (NIR - RED))) / 2
        NDWI <- (GRE - NIR)/(GRE + NIR)
        GNDVI <- (NIR - GRE)/(NIR + GRE)
        OSAVI <- (NIR - RED) / (NIR + RED + 0.16)
        ARVI <- (NIR - (2 * RED) + BLUE) / (NIR + (2 * RED) + BLUE)
        SIPI <- (NIR - BLUE) / (NIR - RED)
        CGI <- (NIR/GRE) - 1
        RVI <- (NIR/RED)
        CVI <- (NIR*RED/(GRE*GRE))
        CCCI <- ((NIR / RED) - 1) / ((NIR / RED) + 1 + BLUE)
        RECI <- (NIR / RED) - 1
        SVI <- NIR / (RED + GRE)
        SSI <- (NIR / RED) - 1
        SCI <- (NIR + RED)/(NIR * RED)
        SMI <- (NIR / RED) ** 0.5
        cal_indices_NIR <- stack(NDVI,EVI,SAVI,MSAVI,NDWI,GNDVI,OSAVI,ARVI,
                             SIPI,CGI,RVI,CVI,CCCI,RECI,SVI,SSI,SCI,SMI)
    }
    return(cal_indices_NIR)
}
```

```r
# Utilizar la funcion cal_indices para indices espectrales
Sent2_Indices_NIR <- cal_indices_NIR(blue, green, red, nir, indices = "indices total")
names(Sent2_Indices_NIR)
```
