## Resamplear pixcel

```r
# Resamplear pixcel
MDE1 <- resample(MDE, Sent2)
# Recortar con el raster para obtener la misma dimension
MDE1 <- crop(MDE1, Sent2)
```

```r
dim(MDE1)
```

```r
dim(Sent2)
```

## Determinacion parametros topogrficos

```r
# Determinacion Pendiente
slope = terrain(MDE1, opt= "slope", unit= "degrees")
# Determinacion Aspecto
aspect = terrain(MDE1, opt= "aspect", unit = "degrees")
# Determinacion TPI: ndice de posicin topogrfica
TPI = terrain(MDE1, opt= "TPI")
# Determinacion TRI: ndice de rugosidad del terreno
TRI = terrain(MDE1, opt= "TRI")
# Determinacion Direccin del flujo de agua - flowdir
flowdir = terrain(MDE1, opt= "flowdir")
```

```r
Topografico = stack(MDE1, slope, aspect, TPI, TRI, flowdir)
```

```r
plot(slope, main= "Pendiente para Zona [grados]" , col= topo.colors(25, alpha = 0.7))
```

```r
plot(aspect, main= "Aspecto para Zona [grados]", col= rainbow(25, alpha = 0.7))
```

```r
plot(TPI, main= "TPI: ndice de posicin topogrfica")
```

```r
plot(TRI , main= "TRI: ndice de rugosidad del terreno")
```

```r
plot(flowdir , main= "Direccion de flujo")
```

```r
carpeta = paste(getwd(),"/","Bandas",sep = "")
carpeta
```

```r
suppressWarnings(dir.create(carpeta))
```

```r
MDE_name <- paste(carpeta, "/S2_45_MDE.tif",sep='')
Slope_name <- paste(carpeta, "/S2_46_slope.tif",sep='')
Aspect_name <- paste(carpeta, "/S2_47_aspect.tif",sep='')
TPI_name <- paste(carpeta, "/S2_48_TPI.tif",sep='')
TRI_name <- paste(carpeta, "/S2_49_TRI.tif",sep='')
FlowDir_name <- paste(carpeta, "/S2_50_flowdir.tif",sep='')
FlowDir_name
```

```r
writeRaster(MDE1, MDE_name, format="GTiff", overwrite=TRUE)
writeRaster(slope, Slope_name, format="GTiff", overwrite=TRUE)
writeRaster(aspect, Aspect_name, format="GTiff", overwrite=TRUE)
writeRaster(TPI, TPI_name, format="GTiff", overwrite=TRUE)
writeRaster(TRI, TRI_name, format="GTiff", overwrite=TRUE)
writeRaster(flowdir, FlowDir_name, format="GTiff", overwrite=TRUE)
```
