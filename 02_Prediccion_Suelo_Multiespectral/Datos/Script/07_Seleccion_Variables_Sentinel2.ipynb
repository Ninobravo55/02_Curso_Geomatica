## Paquete virtualspecies
Esta funci√≥n analiza la correlaci√≥n entre las variables de la pila proporcionada de variables ambientales (utilizando la R de Pearson), y puede devolver un vector que contiene nombres de variables que no son colineales, o una lista que contiene variables de agrupaci√≥n seg√∫n su grado de colinealidad.

```r
#install.packages("virtualspecies",lib = r"(C:\Program Files\R\R-4.2.3\library)")
```

```r
library(virtualspecies)
library(terra)
```

```r
# Convertir las bandas a data.frame
dtf <- as.data.frame(RGB)
names(dtf)
```

```r
# Eliminar valor NA
dtf1 = dtf[!is.na(dtf$S2_01_B1),]
```

```r
dtf2 = dtf1[!is.na(dtf1$S2_46_slope),]
```

```r
dato_correlacion <- cor(dtf2)
dato_correlacion
```

```r
df_cor <- as.data.frame(dato_correlacion)
```

```r
class(df_cor)
```

```r
library(readr)
```

```r
write_csv(df_cor, path = "Variables_correlacion.csv", col_names = TRUE)
```

```r
library(corrplot)
```

```r
#corrplot(dato_correlacion, main = "Matriz de correlaci√≥n")
corrplot(dato_correlacion,
         type="upper",
         tl.col="black",
         tl.srt = 15,
         pch.col = "blue",
         insig = "p-value",
         sig.level = -1,
         col = terrain.colors(100))
```

```r
bandas_reduce <- removeCollinearity(RGB, # Variables ambientales
                                    multicollinearity.cutoff = 0.95, # El l√≠mite de correlaci√≥n por encima del cual se agrupan las variables.
                                    select.variables = T, # entonces la funci√≥n elegir√° una variable entre cada grupo
                                    sample.points = T, # Si tiene un archivo r√°ster grande, use este par√°metro para muestrear una cantidad de puntos igual
                                    plot = T, # Se trazar√° la clasificaci√≥n ascendente jer√°rquica utilizada para agrupar variables.
                                    method = "pearson")
```

```r
# Te devuelve las variables reducidas
bandas_reduce
```

```r
length(bandas_reduce)
```

## Seleccion de variables segun cluster

```r
bandas_select <- subset(RGB, bandas_reduce)
```

```r
# Crear un for
for (i in 1:dim(bandas_select)[3]){
    # Mantiene el nombre de archivo original
    banda <- bandas_select[[i]]
    name <- names(bandas_select[[i]])
    tif_name <- paste(out_dir,"/", name,".tif",sep='')
    # Exportar 
    writeRaster(banda,tif_name, format="Gtiff", overwrite=TRUE)
}
```
