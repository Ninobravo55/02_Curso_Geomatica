# Uso de librerias forestales

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Generar datos simulados
np.random.seed(0)
n_samples = 100
diametro = np.random.uniform(10, 50, n_samples).reshape(-1, 1)  # Diámetro de los árboles en cm
altura = np.random.uniform(2, 20, n_samples).reshape(-1, 1)     # Altura de los árboles en metros
biomasa_real = 0.2 * diametro + 0.5 * altura + np.random.normal(0, 5, n_samples).reshape(-1, 1)  # Biomasa en kg
```

```python
# Visualizar los datos
plt.figure(figsize=(8, 5))
plt.scatter(diametro, biomasa_real, color='blue', label='Datos reales')
plt.xlabel('Diámetro (cm)')
plt.ylabel('Biomasa (kg)')
plt.title('Relación entre el diámetro y la biomasa de los árboles')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Dividir los datos en conjunto de entrenamiento y conjunto de prueba
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(np.concatenate((diametro, altura), axis=1), biomasa_real, test_size=0.2, random_state=42)
```

```python
# Entrenar el modelo de regresión lineal
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Evaluar el modelo con el conjunto de prueba
y_pred = model.predict(X_test)
```

```python
# Visualizar los resultados
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='red')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=3)
plt.xlabel('Biomasa real (kg)')
plt.ylabel('Biomasa predicha (kg)')
plt.title('Biomasa real vs. Biomasa predicha')
plt.grid(True)
plt.show()
```

```python
# Mostrar los coeficientes del modelo
print('Coeficientes del modelo:')
print('  - Intercept:', model.intercept_)
print('  - Coefficients:', model.coef_)
```
