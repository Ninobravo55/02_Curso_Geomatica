# Crear una zona de rectangulo del Ã¡rea interes

```python
import os, glob
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
os.chdir(r"C:\Users\GEOMATICA\Curso_Geomatica\06_BGR_Geomatica\Datos\Shp")
```

```python
os.getcwd()
```

```python
## Lista zip
lista_shp = glob.glob("*.shp")
lista_shp
```

```python
shp = os.getcwd() + "\\" + lista_shp[0]
shp
```

```python
Vuelo_shp = gpd.read_file(shp)
```

```python
Vuelo_shp.head(3)
```

```python
Vuelo_shp.plot(color='none', edgecolor='blue', linestyle='dashed')
```

```python
envolvente = Vuelo_shp.envelope
```

```python
shp_gdp_500 = envolvente.buffer(500, join_style=2)
```

```python
# Crear una figura y ejes
fig, ax = plt.subplots()
# Visualizar la shapefile irregular sin color de fondo y con lÃ­neas punteadas
Vuelo_shp.plot(ax=ax, color='none', edgecolor='blue', linestyle='dashed')
# Visualizar el envolvente sin color de fondo y con lÃ­neas punteadas
shp_gdp_500.plot(ax=ax, color='none', edgecolor='red', linestyle='dashed')
# Configurar tÃ­tulo y etiquetas
ax.set_title('Ã¡rea de interÃ©s (AOI) y su Zona Buffer y envelope')
ax.set_xlabel('Longitud')
ax.set_ylabel('Latitud')
plt.savefig('Grafico_zona_interes.png')
# Mostrar la visualizaciÃ³n
plt.show()
```

```python
os.getcwd()
```

```python
shp_gdp_500.to_file("Zona_buffer_500")
```
